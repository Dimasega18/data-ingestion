```python
df = pd.read_csv('taxi.csv',low_memory=False)
df.head(10)
```

```python
#Cek tipe data
df.info()
```

```python
#connect database
engine = create_engine("postgresql://dimas:d1m45@localhost:5432/db1")
engine.connect()
```

```python
#memasukkan csv ke dalam database

baris = 0

while True:
    try:
        df_new = next(df)
        start = time.time()

        df_new['tpep_dropoff_datetime'] = pd.to_datetime(df_new['tpep_dropoff_datetime'])
        df_new['tpep_pickup_datetime'] = pd.to_datetime(df_new['tpep_pickup_datetime'])
        df_new.to_sql('taxi_data',con=engine,index=False,if_exists='append')
        
        baris += df_new['VendorID'].count()
        end = time.time()
        waktu = end - start
        
        print(f"Data telah di insert dalam waktu {waktu} & total row {baris}")
        time.sleep(3)
        
    except Exception as e:
        print("Terjadi Kesalahan",e)
```
